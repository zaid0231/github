```python

import seaborn as sns
import plotly.express as px
```

```python
df = sns.load_dataset('car_crashes')
```

```python


fig = px.bar(df, x='abbrev',
            y='total',
            color='abbrev',
            title ='Total Car Crashes by State',
            labels = {'abbrev': 'State', 'total': 'Total Crashes'}
)
fig.show()
```

```python
import plotly.express as px
top10 = df.sort_values('ins_premium', ascending=False).head(10)
fig = px.bar(
    top10,
    x='abbrev',
    y='ins_premium',
    title='Top 10 states with highest insurance premium',
    labels={'abbrev': 'State', 'ins_premium': 'Insurance Premium'},
    template='plotly_dark',
    color='ins_premium',
    color_continuous_scale='Agsunset'
)
fig.show()
```

```python
fig = px.scatter(df, x='speeding', y='alcohol', color='abbrev',
                 size='total', title='Speeding vs Alcohol-related crashes',
)
fig.show()
```

```python
df = px.data.tips()
fig = px.sunburst(df, path=['day', 'time', 'sex'], values='total_bill')
fig.show()
```
