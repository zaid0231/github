```python
apps = ["facebook", "instagram", "snapchat", "twitter", "whatsapp", "youtube", "tiktok", "reddit", "pinterest", "linkedin"]
for app in apps:
    print(app, end=" ")
```

Indexing && Slicing

```python
print(apps[3:])
```

```python
print(apps[1:9:2])  # first element
```

LIST METHODS


append()	Adds an element at the end of the list
clear()	    Removes all the elements from the list
copy()	    Returns a copy of the list
extend()	Add the elements of a list (or any iterable), to the end of the current list
insert()	Adds an element at the specified position
pop()	    Removes the element at the specified position
remove()	Removes the first item with the specified value
sort()	    Sorts the list

```python
apps.append("NetFlix") # adds to the end of the list
print(apps)
```

```python
apps.insert(0, "Amazon")
print(apps)
```

```python
apps.pop()  # removes last element
print(apps)
```

```python
apps.remove("snapchat")
print(apps)
```

```python
clone_list = apps.copy()
print(clone_list)
```

```python
clone_list.clear()
print(clone_list)
```

```python
apps.sort()
print(apps)
```

```python
apps_2 = ["uber", "lyft", "airbnb"]
apps.extend(apps_2)
print(apps)
```
