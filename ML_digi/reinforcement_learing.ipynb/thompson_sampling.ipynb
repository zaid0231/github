```python
import random
random.seed(40)
```

```python
n_machines = 5
n_trails = 10
true_probs = [0.1,0.3,0.5,0.2,0.4]
```

```python
successes = [0] * n_machines
failures = [0] * n_machines
```

```python
for t in range(n_trails):
    samples = [random.betavariate(successes[i] + 1, failures[i] + 1) for i in range(n_machines)]
    choice = samples.index(max(samples))
    reward = 1 if random.random() < true_probs[choice] else 0
    if reward :
        successes[choice] += 1
    else:
        failures[choice] += 1
```

```python
estimated_rate = []
for i in range(n_machines):
    total = successes[i] + failures[i]
    est = successes[i] / total if total > 0 else 0.0
    estimated_rate.append(est)
    print(f"Machine{i+1}: Played {total} times, Estimated Success Rate: {est:.2f}")

print("best Machine to play:", estimated_rate.index(max(estimated_rate)) + 1    )
```
