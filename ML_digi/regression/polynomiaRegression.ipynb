```python
fig,(ax1,ax2,ax3,ax4)= plt.subplots(1,4,figsize=(20,5))
ax1.plot(df['x'],color='red',label= 'degree1',marker='o')
ax2.plot(df['X2'],color='blue',label= 'degree2',marker='o')
ax3.plot(df['X3'],color='green',label= 'degree3',marker='o')
ax4.plot(df['X4'],color='purple',label= 'degree4',marker='o')
ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
plt.show()
```

```python
df = pd.read_csv('Position_Salaries.csv')
df.head()
```

```python
plt.scatter(df['Level'],df['Salary'], color='lightcoral')
plt.title('Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show(False)
```

```python
X=df[['Level']]
y=df['Salary']
```

```python
regressor=LinearRegression()
regressor.fit(X,y)
```

```python
pr=PolynomialFeatures(degree=5)
X_poly=pr.fit_transform(X)
polyReg=LinearRegression()
polyReg.fit(X_poly,y)
```

```python
y_pred_lr=regressor.predict(X)
y_pred_poly=polyReg.predict(X_poly)
```
