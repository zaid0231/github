```python


import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python

from xgboost import XGBClassifier
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.isnull().sum()
```

```python
X=df.drop("Outcome", axis=1)
y=df["Outcome"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0,stratify=y)
```

```python
xgb = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=0)
xgb.fit(X_train, y_train)
```

```python
y_pred = xgb.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
print("train class distribution:\n", y_train.value_counts(normalize=True))
print("test class distribution:\n", y_test.value_counts(normalize=True))
```

stratify=y #is ensures that the training and testing sets have the same proportion of classes (or labels) as the original dataset

learning_rate #  This parameter controls the step size at each iteration while moving toward a minimum of the loss function. A smaller learning rate makes the model more robust to overfitting but requires more trees (n_estimators) to converge.
