```python
import math
import random
```

```python
n_machines = 3
n_trails = 10
```

```python
count = [0] * n_machines
reward = [0] * n_machines
```

```python
true_rewards = [0.2, 0.5, 0.8]
```

```python
for t in range(1, n_trails+1):
    if 0 in count:
        choice = count.index(0)
    else:
        ucb_values = [
            reward[i]/count[i]+ math.sqrt(2*math.log(t)/count[i])
            for i in range(n_machines)
        ]
        arm=ucb_values.index(max(ucb_values))
        count[arm] += 1
        reward[arm] += reward
```

```python
for t in range(n_machines):
    print(f"Machine{t+1}: Played {count[t]} times, Estimated win rate: {reward[t]/count[t]:.2f}")
```
