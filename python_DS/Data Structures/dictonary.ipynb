```python
# another way to create a dictionary
student = dict(name = "Zaid", age = 21, Branch = ["DataScience", "ArtificialIntelligence"], sem = 7, Course = "B.Tech")
print(student)  
```

```python
# taking input from user and adding it to dictionary
student["phone"] = input("Enter your phone number: ")
print(student)
```

```python
# accessing elements in a dictionary


#method 1
print(student["name"])


#method 2
print(student.get("age"))
```

```python
student.keys()
```

```python
student.values()
```

```python
student.items()
```

```python
#loop 
for k in student.items():
    print(k)
```

```python
for k, v in student.items():
    print(f"{k} : {v}")
```

```python
# nested dictionary

company = {
    "employee1": {"name": "Alice", "age": 30, "Department": 'HR' },
    "employee2": {"name": "Bob", "age": 25, 'Department': 'IT' },
    "employee3": {"name": "Charlie", "age": 28, 'Department': 'Finance' }
}
```

```python
print(company)  
```

```python
company['employee4'] = {"name": "David", "age": 35, 'Department': 'Marketing' }
print(company)
```

```python
company['employee2']['age']
```
