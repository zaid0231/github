```python
from sqlalchemy import create_engine, Column, Integer, String  # type: ignore
from sqlalchemy.orm import declarative_base, sessionmaker  # type: ignore
```

- create engine : It is used to make connention form the database
- column : I defines columns in database
- Integer and string :  datatype of values in database

```python
engine = create_engine('sqlite:///user.db') # create engine : It is used to make connention form the database
```

```python
Base = declarative_base() # it is used to create classes which are mapped to database tables
```

```python
class User(Base):
    __tablename__ = 'users'
    id = Column(Integer, primary_key=True)
    name = Column(String)
    password = Column(Integer)
```

```python
Base.metadata.create_all(engine) # it creates the tables in the database
```

```python
Session = sessionmaker(bind=engine)
session = Session()
```

Till here all steps are mandatory

Add data

```python
new_user = User(name="john", password = "abrakadabra")
session.add(new_user)
session.commit()
```

```python
new_user = User(name="Alice", password = "pass")
session.add(new_user)
session.commit()
```

```python
# retrieving data
users = session.query(User).all()
for user in users:
    print(f"ID: {user.id}, Name: {user.name}, Password: {user.password}")
```

```python
#Delete data
users = session.query(User).filter_by(id=1).first()
session.delete(users)
session.commit()
```

```python
# retrieving data
users = session.query(User).all()
for user in users:
    print(f"ID: {user.id}, Name: {user.name}, Password: {user.password}")
```
