SPECIAL DATATYPES 

```python
#LIST {represented with square brackets}


my_list = [1, 2, 3, 4, 5]
vals = [1,2.2,'python']
colors = ['red', 'green', 'blue']
print(colors[0])
```

```python
#TUPLE {represented with round brackets}


coordinates = (10.0, 20.0)
print(coordinates[0])
```

```python
# SET {represented with curly brackets}
# NO DUPLICATES ALLOWED ,, VALUES ARE UNORDERED
# REMOVE DUPLICATES FROM A LIST AUTOMATICALLY

my_set = {1, 2, 3, 3,4, 5}
print(my_set)
```

```python
#DICTIONARY {represented with curly brackets, key-value pairs}

user= {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

print(user["name"])  # Accessing value by key
print(user.get("age"))  # Another way to access value by key
```
